# EDA-Univariate Analysis

```python
for col in num_cols:
  # Skip plotting if the column is entirely null
  if data[col].isnull().all():
    print(f"Skipping plotting for '{col}' column as it contains only null values.")
    continue

  print(col)
  print('Skew:',round(data[col].skew(),2))
  plt.figure(figsize=(15,4))
  plt.subplot(1,2,1)
  data[col].hist(grid=False)
  plt.ylabel('Count')
  plt.subplot(1,2,2)
  sns.boxplot(x=data[col])
  plt.show()
```

```python
# Create subplots (3 rows, 2 columns)
fig, axes = plt.subplots(3, 2, figsize=(18, 18))

fig.suptitle('Bar plot for all categorical variables in the dataset', fontsize=16)

# 1. Fuel
sns.countplot(ax=axes[0, 0], x='fuel', data=data, color='blue',
              order=data['fuel'].value_counts().index)

# 2. Transmission
sns.countplot(ax=axes[0, 1], x='transmission', data=data, color='blue',
              order=data['transmission'].value_counts().index)

# 3. Condition (replacing Owner_Type)
sns.countplot(ax=axes[1, 0], x='condition', data=data, color='blue',
              order=data['condition'].value_counts().index)

# 4. Type (replacing Location)
sns.countplot(ax=axes[1, 1], x='type', data=data, color='blue',
              order=data['type'].value_counts().index)

# 5. Brand (top 20)
sns.countplot(ax=axes[2, 0], x='Brand', data=data, color='blue',
              order=data['Brand'].value_counts().head(20).index)

# 6. Model (top 20)
sns.countplot(ax=axes[2, 1], x='Model', data=data, color='blue',
              order=data['Model'].value_counts().head(20).index)

# Rotate labels for readability
axes[1][1].tick_params(labelrotation=45)
axes[2][0].tick_params(labelrotation=90)
axes[2][1].tick_params(labelrotation=90)

plt.tight_layout()
plt.show()
```

# **Data Transformation**

```python
def log_transform(data,col_list):
  for colname in col_list:
    # Using np.log1p for robustness, as log(0) is undefined and original condition (data[colname]==1.0).all() is very specific
    # and likely not intended for general log transformation.
    # Also fill NaNs with 0 before taking log if that's the desired behavior
    data[colname + '_log'] = np.log1p(data[colname].fillna(0))
```

```python
log_transform(data,['odometer','price'])
```

```python
#log transformation for the feature 'odometer'
sns.distplot(data['odometer_log'],axlabel="odometer_log")
```

# **Eda Bivariate Analysis**

```python
plt.figure(figsize=(13,17))
sns.pairplot(data=data.drop(columns=['odometer','price']))
plt.show()
```

```python


fig, axarr = plt.subplots(4, 2, figsize=(12, 18))

# Create Car_Age first
data['Car_Age'] = data['Current_year'] - data['year']

data.groupby('region')['price_log'].mean().sort_values(ascending=False).plot.bar(ax=axarr[0][0], fontsize=12)
axarr[0][0].set_title('Region vs Price')

data.groupby('transmission')['price_log'].mean().sort_values(ascending=False).plot.bar(ax=axarr[0][1], fontsize=12)
axarr[0][1].set_title('Transmission vs Price')

data.groupby('fuel')['price_log'].mean().sort_values(ascending=False).plot.bar(ax=axarr[1][0], fontsize=12)
axarr[1][0].set_title('Fuel Type vs Price')

data.groupby('condition')['price_log'].mean().sort_values(ascending=False).plot.bar(ax=axarr[1][1], fontsize=12)
axarr[1][1].set_title('Condition vs Price')

data.groupby('Brand')['price_log'].mean().sort_values(ascending=False).head(20).plot.bar(ax=axarr[2][0], fontsize=12)
axarr[2][0].set_title('Brand vs Price')

data.groupby('Model')['price_log'].mean().sort_values(ascending=False).head(20).plot.bar(ax=axarr[2][1], fontsize=12)
axarr[2][1].set_title('Model vs Price')

data.groupby('type')['price_log'].mean().sort_values(ascending=False).plot.bar(ax=axarr[3][0], fontsize=12)
axarr[3][0].set_title('Type vs Price')

data.groupby('Car_Age')['price_log'].mean().sort_values(ascending=False).plot.bar(ax=axarr[3][1], fontsize=12)
axarr[3][1].set_title('Car Age vs Price')

plt.subplots_adjust(hspace=0.5)
sns.despine()
```
